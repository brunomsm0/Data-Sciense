```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel('golf_df.xlsx')
dados.head()
```

```python
x = pd.get_dummies(dados[["clima","temp","umidade","vento"]])
x.head()
```

```python
y = dados[["jogar"]]
y
```

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
```

```python
arvore = arvore.fit(x, y)
```

```python
modelo = arvore.fit(x, y)
```

```python
modelo
tree.DecisionTreeClassifier()
```

```python
fig = plt.figure(figsize=(40, 15))

fig = tree.plot_tree(modelo, feature_names=x.columns, filled=True)
```
